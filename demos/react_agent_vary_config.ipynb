```python
from agents.react_lg_agent import create_agent
```

```python
# Configuration 1: Conservative
config1 = {
    "model": "gpt-4",
    "temperature": 0.1,
    "top_p": 0.3
}

agent1 = create_agent("casual", **config1)
test_question = "Hello! What services does EduZen offer?"

print("=== Configuration 1: Conservative ===")
print(f"Config: {agent1.get_config()}")
print("\n" + "="*50)

response1, reasoning1 = agent1.chat(test_question, thread_id="config1")
print(f"Response: {response1}")
print(f"\nReasoning Steps: {reasoning1}")
```

```python
# Configuration 2: Balanced
config2 = {
    "model": "gpt-4",
    "temperature": 0.7,
    "top_p": 0.8
}

agent2 = create_agent("casual", **config2)

print("=== Configuration 2: Balanced ===")
print(f"Config: {agent2.get_config()}")
print("\n" + "="*50)

response2, reasoning2 = agent2.chat(test_question, thread_id="config2")
print(f"Response: {response2}")
print(f"\nReasoning Steps: {reasoning2}")
```

```python
# Configuration 3: Creative
config3 = {
    "model": "gpt-4",
    "temperature": 1.2,
    "top_p": 0.95
}

agent3 = create_agent("casual", **config3)

print("=== Configuration 3: Creative ===")
print(f"Config: {agent3.get_config()}")
print("\n" + "="*50)

response3, reasoning3 = agent3.chat(test_question, thread_id="config3")
print(f"Response: {response3}")
print(f"\nReasoning Steps: {reasoning3}")
```

The three modes where consistent in the message delivered with slight changes in the vocabulary used.
For our application using the conservative low creativity mode is most suitable.
